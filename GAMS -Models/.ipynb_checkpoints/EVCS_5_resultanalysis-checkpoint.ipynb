```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
import glob
```

```python
all_files = glob.glob('EVCS_5_results.db')

storage = pd.DataFrame()

for filename in glob.glob('EVCS_5_results.db'):
    con = sqlite3.connect(filename)
    grid_power = pd.read_sql_query("SELECT * FROM var_grid_power", con)
    el_cost = pd.read_sql_query("SELECT * FROM var_electricity_cost", con)
    #ESS_size = pd.read_sql_query('SELECT * FROM var_ESS_size', con)
    ESS_level = pd.read_sql_query('SELECT * FROM var_ESS_level', con)
    charging = pd.read_sql_query('SELECT * FROM var_grid2ess_power', con)
    discharging = pd.read_sql_query('SELECT * FROM var_ess2load_power', con)
    grid2load = pd.read_sql_query('SELECT * FROM var_grid2load_power', con)
    #ess_rated = pd.read_sql_query('SELECT * FROM var_ess_rated_power', con) WHY NOT???
    
    con.close()
```
