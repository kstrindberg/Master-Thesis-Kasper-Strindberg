In this model i have managed to implement a hierarchy for the power flow. The soft grid limit is set at 180. There
is a penalty for each timestep the battery is not recharged, and the power used above the soft grid limit is twice 
as costly as the power below. This may not be super realistic, but it gives the desired power flow

The final touches will be ESS efficiency and separate the power tariff

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
import glob
```

```python
all_files = glob.glob('EVCS_14_results.db')
print(all_files)

storage = pd.DataFrame()

for filename in glob.glob('EVCS_14_results.db'):
    con = sqlite3.connect(filename)
    
    grid_power = pd.read_sql_query("SELECT * FROM var_grid_power", con)
    grid_extra_power = pd.read_sql_query("SELECT * FROM var_grid_ex_power", con)
    grid_total_power = pd.read_sql_query("SELECT * FROM var_grid_total_power", con)
    
    grid_energy = pd.read_sql_query("SELECT * FROM var_grid_energy", con)
    grid_extra_energy = pd.read_sql_query("SELECT * FROM var_grid_ex_energy", con)
    
    ESS_level = pd.read_sql_query('SELECT * FROM var_ESS_level', con)
    ESS_power = pd.read_sql_query('SELECT * FROM var_ESS_power', con)
    
    SOC_pen = pd.read_sql_query('SELECT * FROM var_SOC_pen', con)

    tiss = pd.read_sql_query('SELECT * FROM scalarvariables', con)
    

    
    
    con.close()

# Other results    
ESS_size = 141.1970
system_cost = 196274
energy_tax = 3585.8921
grid_tariff = 64541.6237
electricity_cost = 93787.8902
ESS_cost = 31063.3398


# Demand
demand = pd.read_csv('demand_analysis.tsv', sep = ',')
```
