```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
import glob
```

```python
all_files = glob.glob('EVCS_6_results.db')

storage = pd.DataFrame()

for filename in glob.glob('EVCS_6_results.db'):
    con = sqlite3.connect(filename)
    grid_power = pd.read_sql_query("SELECT * FROM var_grid_power", con)
    grid_energy = pd.read_sql_query("SELECT * FROM var_grid_energy", con)
    el_cost = pd.read_sql_query("SELECT * FROM var_electricity_cost", con)
    #ESS_size = pd.read_sql_query('SELECT * FROM var_ESS_size', con)
    ESS_level = pd.read_sql_query('SELECT * FROM var_ESS_level', con)
    charging = pd.read_sql_query('SELECT * FROM var_grid2ess_power', con)
    discharging = pd.read_sql_query('SELECT * FROM var_ess2load_power', con)
    grid2load = pd.read_sql_query('SELECT * FROM var_grid2load_power', con)
    #ess_rated = pd.read_sql_query('SELECT * FROM var_ess_rated_power', con) WHY NOT???
    
    con.close()
```

```python
plt.plot(charging['level'], label = 'charging [kW]')
plt.plot(discharging['level'], label = 'discharging [kW]')
plt.plot(ESS_level['level'], label = 'Energy in battery [kWh]')
plt.legend()
```

```python
cha = charging['level'].tolist()
dis = discharging['level'].tolist()
lol = []

for i in range(len(cha)):
    if cha[i] > 0 and dis[i] > 0:
        lol.append(i)
    else:
        pass
    
print(str(len(lol)) + ' timesteps where battery is both charging and discharging')
```
