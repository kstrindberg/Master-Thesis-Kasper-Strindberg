```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
df1 = glob.glob('kong_*.csv')#, sep=';', decimal=',')
#df1['DateTime'] = pd.to_datetime(df1['DateTime'])
#df1 = df1.set_index('DateTime', drop=False)

files = []

for filename in glob.glob('kong_**.csv'):
    files.append(pd.read_csv(filename, sep=';', decimal=','))
    
df = pd.concat(files, ignore_index = False)
df['DateTime'] = pd.to_datetime(df['DateTime'])
df = df.set_index('DateTime', drop=False)
```

```python
time = np.arange(0,34542,1)
ind = []

for i in range(len(time)):
    ind.append('m' + str(time[i]))
```
